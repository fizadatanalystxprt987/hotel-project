# canceled count

```python
canceled_percentage= df['is_canceled'].value_counts(normalize=True)
canceled_percentage
```

```python
plt.figure(figsize=(5,4))
plt.title('Reservation_status')
plt.bar(['no_canceled','canceled'],df['is_canceled'].value_counts(),edgecolor='black',width=0.8)
plt.show()
```

# cancellation rate with respect to hotels

```python
plt.figure(figsize=(8,4))
ax_1=sb.countplot(x='hotel',hue='is_canceled',data=df,palette='rainbow')
legend_labels,_=ax_1.get_legend_handles_labels()
ax_1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status with respect to Hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('number of resrvations')
plt.legend(['no_canceled','canceled'])
plt.show()
```
